# Problem 1

Consider a test for a disease that has a true positive rate of 0.99 and a false positive rate of 0.001  Assume the disease has a prevalence of 0.0001.   What is the probability that a person receiving a positive test result does not have the disease?

We want to compute $P(not sick|positive)$, given $P(positive|sick)=0.99$, $P(positive|not sick)=0.001$, and $P(sick)=0.0001$. By Bayes' Theorem,
$$P(not sick|positive) = P(positive|not sick)\frac{P(not sick)}{P(positive)}$$
$$= 0.001\frac{0.9999}{.99+0.001}$$
$$\approx 0.001009$$

So the probability that a person receiving a positive test result does not have the disease is just over 0.1%

# Problem 2
Generate a realization of the chi-squared distribution for 2 degrees of freedom.

(1) Present a plot of a realization of samples drawn from this distribution

```python
import numpy as np
import matplotlib.pyplot as plt
N=10**6
samples = np.random.chisquare(2,N)
plt.plot(samples,',')
plt.show()
```

(2) Plot a histogram of this realization, choose the number of samples to be sufficiently high that the shape of the distribution can be seen with limited noise.

```python
numBins=100
plt.hist(samples,bins=numBins)
plt.show()
```

(3) Compute the mean, standard deviation, skewness, and kurtosis 

```python
mean = np.mean(samples)
stdev = np.std(samples)
meansub = samples - mean
skewness = sum(meansub**3)/N-1
kurtosis = sum(meansub**4)/N-1
print("mean: ", mean)
print("standard deviation: ",stdev)
print("skewness: ",skewness)
print("kurtosis: ",kurtosis)
```
